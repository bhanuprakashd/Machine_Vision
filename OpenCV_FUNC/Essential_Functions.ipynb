```python
methods=[cv2.INTER_LINEAR,cv2.INTER_NEAREST,cv2.INTER_AREA,cv2.INTER_CUBIC,cv2.INTER_LANCZOS4]
labels=['INTER_LINEAR','INTER_NEAREST','INTER_AREA','INTER_CUBIC','INTER_LANCZOS4']
```

```python
#Using Loop
for i in range(5):
    smaller_image = cv2.resize(image,(100,100),methods[i])
    i=i+1
    plt.subplot(2,3,i)
    plt.title(labels[i-1])
    plt.imshow(smaller_image)
    
```

```python
#Rotating Images
rows,cols = image.shape[:2] 
#(col/2,rows/2) is the center of rotation for the image 
# M is the cordinates of the center 
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1) 
rtd = cv2.warpAffine(image,M,(cols,rows)) 
plt.imshow(rtd)
```

```python
#Using loop
angles=[0,45,90,180,270,360]
for i in range(6):
    M = cv2.getRotationMatrix2D((cols/2,rows/2),angles[i],1) 
    rtd = cv2.warpAffine(image,M,(cols,rows)) 
    i=i+1
    plt.subplot(2,3,i)
    plt.title("angle" +" "+ str(angles[i-1]))
    plt.imshow(rtd)
```

```python
#Translation
#Shifting the image 100 pixels in both dimensions
M=np.float32([[1,0,-100],[0,1,-100]])
tsl=cv2.warpAffine(image,M,(cols,rows))
plt.imshow(tsl)
```

```python
#Thresholding

gray_image = cv2.imread('index.png',0)
plt.title("Gray_scale Image")
plt.imshow(gray_image,'gray')
```

```python
methods = [cv2.THRESH_BINARY,cv2.THRESH_BINARY_INV,cv2.THRESH_TRUNC,cv2.THRESH_TOZERO,cv2.THRESH_TOZERO_INV]
names = ['BINARY','THRESH_BINARY_INV','THRESH_TRUNC','THRESH_TOZERO','THRESH_TOZERO_INV']
```

```python
for i in range(5):
    ret,threshold = cv2.threshold(gray_image,127,255,methods[i])
    plt.subplot(2,3,i+1)
    plt.title(names[i])
    plt.imshow(threshold,'gray')
plt.subplot(2,3,6)
plt.title("Original image")
plt.imshow(gray_image,'gray')
plt.show()
```

```python
#Adaptive Thresholding
gray_image=cv2.imread('index.png',0)
plt.subplot(2,2,1)
plt.title('Original_image')
plt.imshow(gray_image,'gray')
ret,threshold_global=cv2.threshold(gray_image,127,255,cv2.THRESH_BINARY)
plt.subplot(2,2,2)
plt.title('Global_Thresholding')
plt.imshow(threshold_global,'gray')
methods=[cv2.ADAPTIVE_THRESH_MEAN_C,cv2.ADAPTIVE_THRESH_GAUSSIAN_C]
names=['Adaptive Mean Threshold','Adaptive Gaussian Thresholding']
for i in range(2):
    thresh_adaptive=cv2.adaptiveThreshold(gray_image,255,methods[i],cv2.THRESH_BINARY,11,2)
    plt.subplot(2,2,3+i)
    plt.title(names[i])
    plt.imshow(thresh_adaptive,'gray')
plt.show()
```
